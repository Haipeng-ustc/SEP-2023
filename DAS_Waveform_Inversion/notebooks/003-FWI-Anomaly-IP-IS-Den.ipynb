### FWI using horizontal DAS cable.
The IP, IS, and Density parameterization is used.

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.io as sio
```

```python
font_size = 10
params = {
    'image.cmap': 'seismic',
    'axes.grid': False,
    'savefig.dpi': 300,   # to adjust notebook inline plot size
    'axes.labelsize': font_size, # fontsize for x and y labels (was 10)
    'axes.titlesize': font_size,
    'font.size': font_size, 
    'legend.fontsize': font_size,
    'xtick.labelsize': font_size,
    'ytick.labelsize': font_size
}
plt.rcParams.update(params)
```

```python
# set up model size
nx = 201
nz = 101
dx = 20.0
dz = 20.0
dt = 0.002
nt = 1501
x = np.arange(0, nx * dx, dx)
z = np.arange(0, nz * dz, dz)

# true model
vp = np.ones((nx, nz)) * 4000.0
vs = np.ones((nx, nz)) * 4000.0 / 1.732
rho = np.ones((nx, nz)) * 2500.0
IP = vp * rho
IS = vs * rho

# set anomaly scale
scale_IP = IP[0,0] * 0.025
scale_IS = IS[0,0] * 0.025
scale_rho= 50

IP[42:58, 42:58] += scale_IP
IS[92:108, 42:58] -= scale_IS
rho[142:158, 42:58] += scale_rho

# true vp and vs
vp = IP/rho
vs = IS/rho

# initial model
vp_init = np.ones((nx, nz)) * 4000.0
vs_init = np.ones((nx, nz)) * 4000.0 / 1.732
rho_init = np.ones((nx, nz)) * 2500.0
IP_init = vp_init * rho_init
IS_init = vs_init * rho_init

# set source and receiver
ind_src_x =   np.arange(10, nx-10, 10).astype(int)
ind_src_z = 1*np.ones(ind_src_x.shape[0]).astype(int)
ind_rec_x =   np.arange(10, nx-10).astype(int)
ind_rec_z =95*np.ones(ind_rec_x.shape[0]).astype(int)
```

```python
# save model
path_save = './Models/'
np.savetxt(path_save+"Anomaly_P-WAVE_VELOCITY_101_201_IP_IS_Den.txt", vp.T)
np.savetxt(path_save+"Anomaly_S-WAVE_VELOCITY_101_201_IP_IS_Den.txt", vs.T)
np.savetxt(path_save+"Anomaly_DENSITY_101_201_IP_IS_Den.txt",        rho.T)
np.savetxt(path_save+"Anomaly_P-WAVE_VELOCITY_101_201_IP_IS_Den_init.txt", vp_init.T)
np.savetxt(path_save+"Anomaly_S-WAVE_VELOCITY_101_201_IP_IS_Den_init.txt", vs_init.T)
np.savetxt(path_save+"Anomaly_DENSITY_101_201_IP_IS_Den_init.txt",        rho_init.T)
```

```python
# plot the results

ite = 99
work_path = '/scratch/users/haipeng/Anomaly-IP-IS-Den'
IP_fwi  = sio.loadmat(work_path + f'/Results/IP{ite}.mat')['IP'].T
IS_fwi  = sio.loadmat(work_path + f'/Results/IS{ite}.mat')['IS'].T
rho_fwi = sio.loadmat(work_path + f'/Results/Den{ite}.mat')['Den'].T

fig = plt.figure(figsize=(12, 6))
data = [IP, IS, rho, IP-IP_init, IS-IS_init, rho-rho_init, IP_fwi*1e3-IP_init, IS_fwi*1e3-IS_init, rho_fwi-rho_init]
for i, data in enumerate(data):
    
    ax = fig.add_subplot(3, 3, (i+1))
    if i < 3:
        im = ax.pcolor(x/1000, z/1000, data.T, cmap ='jet')
    else: 
        im = ax.pcolor(x/1000, z/1000, data.T, cmap ='seismic', vmin = -abs(data).max(), vmax= abs(data).max())

    if i == 0:
        ax.plot(ind_rec_x * dx/1000, ind_rec_z * dz/1000, 'w-', linewidth=2.5)
        ax.scatter(ind_src_x * dx/1000, ind_src_z * dz/1000, color='r', marker = '*')

    plt.xticks(fontsize=14)
    plt.yticks(fontsize=14)
    plt.gca().invert_yaxis()
    colorbar = fig.colorbar(im, shrink=0.8)
    colorbar.ax.tick_params(labelsize=10)
    
    ax.set_aspect('equal')
    if i > 5:
        plt.xlabel('Distance (km)', fontsize=14)
    if i % 3 == 0:
        plt.ylabel('Depth (km)', fontsize=14)

plt.tight_layout()
plt.savefig('./Fig/FWI-Anomaly-IP-IS-Den.png', dpi=300, bbox_inches='tight')
```
